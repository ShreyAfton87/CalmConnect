```python
from PIL import Image
import pytesseract
import os

# Function to extract text from an image
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text

# Function to convert image to grayscale while preserving colored text
def convert_to_grayscale_preserve_text(image_path):
    image = Image.open(image_path)
    # Convert the image to grayscale
    grayscale_image = image.convert('L')
    
    # Use the original image to identify regions with colored text
    original_image = Image.open(image_path)
    grayscale_image = grayscale_image.convert('RGB')

    # Iterate through each pixel to preserve colored text
    for x in range(image.width):
        for y in range(image.height):
            r, g, b = original_image.getpixel((x, y))
            if r + g + b < 750:  # You may need to adjust this threshold based on your specific images
                grayscale_image.putpixel((x, y), (r, g, b))

    return grayscale_image

# Directory containing your images
image_directory = 'C:/workarea2/myPython/myDSproject01-3yr-5sem/Emotion-recognition/output'

# List to store the extracted text and grayscale images
text_and_grayscale_images = []

# Loop through each image in the directory
for filename in os.listdir(image_directory):
    if filename.endswith(('.jpg', '.png', '.jpeg')):
        image_path = os.path.join(image_directory, filename)
        extracted_text = extract_text_from_image(image_path)
        grayscale_image = convert_to_grayscale_preserve_text(image_path)
        text_and_grayscale_images.append((extracted_text, grayscale_image))

# Process or save text and grayscale images as needed
for i, (text, grayscale_image) in enumerate(text_and_grayscale_images):
    grayscale_image.save(f'grayscale_image_{i}.png')  # Save grayscale image
    print(f"Text from Image {i + 1}:\n{text}\n")

# You can now work with the 'text_and_grayscale_images' list as needed
```

```python
from PIL import Image
import pytesseract
import os
from textblob import TextBlob
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

# Function to extract text from an image
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    image = image.convert('L')  # Convert the image to grayscale
    text = pytesseract.image_to_string(image)
    return text

# Directory containing your images
image_directory = 'C:/workarea2/myPython/myDSproject01-3yr-5sem/Emotion-recognition/output'

# List to store the extracted text
extracted_text_list = []

# Loop through images in the directory and extract text
for filename in os.listdir(image_directory):
    if filename.endswith(('.jpg', '.png', '.jpeg')):
        image_path = os.path.join(image_directory, filename)
        extracted_text = extract_text_from_image(image_path)
        extracted_text_list.append(extracted_text)

# Print the list of words
for text in extracted_text_list:
    words = text.split()  # Split the extracted text into words
    for word in words:
        print(word)

# Initialize variables to keep track of sentiment scores
positive_sentiment_score = 0
negative_sentiment_score = 0

# Process each extracted text and calculate sentiment scores
for text in extracted_text_list:
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity

    if sentiment > 0:
        positive_sentiment_score += sentiment
    elif sentiment < 0:
        negative_sentiment_score += sentiment

# Print the sentiment scores
print("Positive Sentiment Score:", positive_sentiment_score)
print("Negative Sentiment Score:", negative_sentiment_score)

#emotions = {'joy': 0, 'sadness': 0, 'anger': 0, 'fear': 0, 'surprise': 0}
emotions = {'angry':0 ,'disgust':0,'scared':0, 'happy':0, 'sad':0, 'surprised':0,'neutral':0}

analyzer = SentimentIntensityAnalyzer()

# Process each extracted text and calculate sentiment scores and emotions
for text in extracted_text_list:
    # Sentiment Analysis using TextBlob
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        positive_sentiment_score += sentiment
    elif sentiment < 0:
        negative_sentiment_score += sentiment

    # Emotional Analysis using NLTK's VADER
    sentiment_dict = analyzer.polarity_scores(text)
    for emotion, score in sentiment_dict.items():
        if emotion in emotions:
            emotions[emotion] += score

print("Emotions:", emotions)

# Visualize the emotions
plt.bar(emotions.keys(), emotions.values())
plt.xlabel('Emotion')
plt.ylabel('Intensity Score')
plt.title('Emotional Analysis')
plt.show()
```
